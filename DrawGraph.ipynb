```python
import re
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def draw_graph_from_input(s: str):
    pairs = re.findall(r'\((\w)\s*,\s*(\d+)\)', s)
    names = [v for v, _ in pairs]
    xs    = [int(x) for _, x in pairs]
    n     = len(names)

    G = nx.DiGraph()
    G.add_nodes_from(names)
    for i, v in enumerate(names):
        x = xs[i] % n
        right = (i + x) % n
        left  = (i - x) % n
        G.add_edge(v, names[right])
        G.add_edge(v, names[left])

    pos = nx.circular_layout(G)
    nx.draw(G, pos, with_labels=True, arrowsize=20, node_size=1000)
    plt.title("Directed Graph from offsets")
    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    raw = input("Enter list of (V,x) pairs: ")
    draw_graph_from_input(raw)
```
